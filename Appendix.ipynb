# Reading the lenses database

Here we show how to read the database of lensed images, and how to generate a statistically equivalent distribution of their events.

Three thresholds to define a high magnification event are available: 0.3, 0.5 and 1.0 $\Delta$m.

For each threshold, in each of the u, g, r, i, z and y bands, the statistics of events can be displayed.

Here we can reproduce all the plots presented in the paper, for each combination of band and threshold.

```python
import nav22
```

```python
band = 'r'
threshold = 0.3

lenses = nav22.DB(band=band, threshold=threshold)
```

```python
lenses.lenses
```

```python
lenses.expectedEvents()
```

```python
lenses.fractionCaustic()
```

```python
lenses.plot_all()
```

```python
lenses.plot_all(cc_only=True)
```

```python
lenses.plot_all(exclude_cc=True)
```
